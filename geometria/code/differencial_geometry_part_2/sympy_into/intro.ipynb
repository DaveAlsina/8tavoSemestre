# Introducción a sympy

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

Creamos variables simbolicas

```python
t = sp.symbols('t')
f = 3*t**2 + 2*t - 1
```

diferenciar

```python
sp.diff(f, t)
```

integrar

```python
sp.integrate(f, t)
```

evaluar

```python
f.subs(t, 2)
```

Evaluamos un dominio y ploteamos

```python
domain = np.linspace(start=-5, stop=5, num=100)

#evaluate f at each point in domain
fn = sp.lambdify(t, f, 'numpy')

plt.plot(domain, fn(domain))
```

Círculo parametrizado

```python
x = sp.cos(t)
y = sp.sin(t)
z = 0 

#vectorial function
alpha = (x, y, z)
alpha
```

```python
delta_alpha = (sp.diff(x, t), sp.diff(y, t), sp.diff(z, t))
delta_alpha
```

```python
domain = np.linspace(start=0, stop=2*np.pi, num=100)
fn2 = sp.lambdify(t, alpha, 'numpy')
fn3 = sp.lambdify(t, delta_alpha, 'numpy')

plt.title('Parametric curve of the unit circle')
plt.plot(fn2(domain)[0], fn2(domain)[1])

#show the tangent lines derivatives
plt.quiver(fn2(domain)[0], fn2(domain)[1], fn3(domain)[0], fn3(domain)[1], scale=3, label='Tangent lines')
plt.legend()
```
