```julia
include("function_aproximation.jl")
using Plots
```

```julia
fn = x -> sin(x)
lower_bound = 0
upper_bound = pi

# Plot the function
plot(fn, lower_bound, upper_bound)
```

```julia
degree = 4
lower_bound = 0.0
upper_bound = Float64(2*pi)

aprox_fn = aproximate_fn_with_polinomial(fn, lower_bound, upper_bound, degree)

# Plot the aproximation
plot(fn, lower_bound, upper_bound)
plot!(aprox_fn, lower_bound, upper_bound)
```

```julia
degree = 2
fn = x -> exp(x)
lower_bound = -1.0
upper_bound = 1.0

aprox_fn1 = aproximate_fn_with_polinomial(fn, lower_bound, upper_bound, degree)
aprox_fn2 = aproximate_fn_with_polinomial(fn, lower_bound, upper_bound, degree+1)

# the original function is dashed and in red
plot(fn, lower_bound, upper_bound, label="fn", color="red", linestyle=:dash)
plot!(aprox_fn1, lower_bound, upper_bound, label="degree 2", color="blue")
plot!(aprox_fn2, lower_bound, upper_bound, label="degree 3", color="green")
```
