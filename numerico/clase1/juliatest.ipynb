```julia
function eulerExp(n, x)

    value = 0 

    for m in 0:n
        nth = (x^m/factorial(big(m)))
        value += nth
    end

    return value 
end
```

```julia
function getError(realValue, calcValue)
    return abs(realValue - calcValue)/realValue
end
```

```julia
x = 0.5
n = 1
nmax = 100
error = 0.05

result = exp(x)
approx = Inf


while (getError(result, approx) > error) && (n < nmax)
    approx = eulerExp(n, x)
    n += 1 
end

println(n)
println("error: $(getError(result, approx))")
```
