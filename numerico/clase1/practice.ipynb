• Desarrollar un código que permita sumar los dígitos de un número entero no negativo. Por ejemplo, la suma de los dígitos del 3433 es 13. El programa debe pedir el número que desea sumar.

```julia
function sum_digits(n::String)::Int64

    """
    Add up the digits inside a number 'n' inputed as a string.

    Parameters
    ----------
    n : String
        Number to be added up.

    Returns
    -------
    ans : Int64
    """

    return sum([parse(Int64, i) for i in n])

end
```

```julia
number = "123091283109283"
sum_digits(number)
```

- Reciba como entrada un número entero $n$ y entregue como salida el n-ésimo número de Fibonacci. 

```julia
function nth_fibonacci(n::Int64)
    """
    Give the nth fibonacci number
    
    Parameters
    ----------
    n: Int64, The index of the fibonacci number

    Returns
    -------
    Int64, The nth fibonacci number
    """

    if n <= 0
        error("n must be positive")
    end

    if n == 1 
        return 0
    elseif n == 2
        return 1
    else
        return nth_fibonacci(n-1) + nth_fibonacci(n-2)
    end
end
```

```julia
function nth_fibonnaci_iterative(n::Int64)::Int64

    if n <= 0
        error("n must be positive")
    end

    if n == 1
        return 0
    end

    if n == 2
        return 1
    end

    fib1 = 0 # fib(n-2)
    fib2 = 1 # fib(n-1)

    for i in 3:n
        fibn = fib1 + fib2
        fib1 = fib2
        fib2 = fibn
    end

    return fib2
end
```

```julia
#don't go over 35, it takes too long
nmax = 35 

println("Testing the two methods to calculate the nth Fibonacci number")

#compares the results the two methods give 
for n in range(1, nmax)
    
    print(".")
    ans1 = nth_fibonacci(n)
    ans2 = nth_fibonnaci_iterative(n)

    if ans1 != ans2
        print("n = $(n)\n")
        error("The two methods don't give the same result, $(ans1) != $(ans2)")
    end
end
```

- Reciba como entrada un número entero $n$ e indique si es o no un número de Fibonacci.

```julia
function is_square(n::Int64)

    root = sqrt(n)

    #very nasty way to check if root is an integer
    if root % 1 == 0
        return true
    else
        return false
    end
end
```

```julia
function is_fibonacci(n::Int64)

    """
    Applies the Binet's formula to determine if a number is a Fibonacci number.

    Parameters
    ----------
    n : Int64, Number to be evaluated.

    Returns
    -------
    Bool, True if n is a Fibonacci number, False otherwise.
    """

    a = 5*n^2 + 4
    b = 5*n^2 - 4

    if  is_square(a) || is_square(b)
        return true
    else
        return false
    end
end
```

```julia
candidate = 69
println("$candidate is Fibonacci? $(is_fibonacci(candidate))")

#checks if the Fibonacci numbers are classified correctly
println("Checking correct classification of the first $nmax Fibonacci numbers...")
for n in range(1, nmax)

    if !is_fibonacci(nth_fibonacci(n))
        println("Error: $n is not classified as a Fibonacci number")
    end
end
```
