```python
import numpy as np
import matplotlib.pyplot as plt

L = 1.0  # Longitud de la barra
alpha = 1.0  # Parámetro de difusión térmica
N = 100  # Número de puntos en la cuadrícula
dx = L / (N - 1)  # Tamaño del paso espacial

# Definir la cuadrícula espacial
x = np.linspace(0, L, N)

# Inicializar la temperatura (puedes elegir una condición inicial específica)
T = np.zeros(N)
T[0] = 100  # Temperatura en el extremo izquierdo
```

```python
A = np.zeros((N, N))
b = np.zeros(N)

for i in range(1, N-1):
    A[i, i-1] = 1
    A[i, i] = -2 - alpha**2 * dx**2
    A[i, i+1] = 1

# Condiciones de contorno (por ejemplo, extremos de la barra a temperatura constante)
A[0, 0] = 1
A[N-1, N-1] = 1

b[0] = T[0]
b[N-1] = T[N-1]
```

```python
T = np.linalg.solve(A, b)
```

```python
plt.plot(x, T)
plt.xlabel('Posición (x)')
plt.ylabel('Temperatura (T)')
plt.title('Distribución de Temperatura en Estado Estacionario')
plt.show()
```
